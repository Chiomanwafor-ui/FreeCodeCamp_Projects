# 1. Scatter Plot Game

#### Show random points, player guesses coordinates.

```python
import matplotlib.pyplot as plt
import random

def scatter_plot_game():
    # Random point coordinates
    x = random.randint(0, 10)
    y = random.randint(0, 10)

    # Create scatter plot
    plt.scatter(x, y, color='red', s=100)
    plt.xlim(0, 10)
    plt.ylim(0, 10)
    plt.grid(True)
    plt.title("Guess the Coordinates of the Point")
    plt.show()

    # Get player guess
    guess_x = int(input("Enter the x-coordinate: "))
    guess_y = int(input("Enter the y-coordinate: "))

    # Check correctness
    if guess_x == x and guess_y == y:
        print("âœ… Correct! Well done.")
    else:
        print({x}, {y})

scatter_plot_game()
```

# 2. Algebra Practice Game

#### Solve 1-step & 2-step random algebra problems.

```python
def algebra_game():
    import operator

    ops = {
        '+': operator.add,
        '-': operator.sub,
        '*': operator.mul
    }

    step_type = random.choice([1, 2])

    if step_type == 1:  # One-step equation: x + b = c
        a = random.randint(-10, 10)
        b = random.randint(-10, 10)
        op = random.choice(list(ops.keys()))
        answer = a
        equation = f"x {op} {b} = {ops[op](a, b)}"

    else:  # Two-step equation: a*x + b = c
        a = random.randint(1, 5)
        b = random.randint(-10, 10)
        x_val = random.randint(-10, 10)
        c = a * x_val + b
        equation = f"{a}x + {b} = {c}"
        answer = x_val

    print("Solve:", equation)
    guess = int(input("Your answer: "))

    if guess == answer:
        print("âœ… Correct!")
    else:
        print({answer})

algebra_game()
```

# 3. Projectile Game


#### Clear a wall by adjusting a, b, c in y = axÂ² + bx + c.

```python
import numpy as np

def projectile_game():
    # Random wall position and height
    wall_x = random.randint(2, 8)
    wall_height = random.randint(3, 10)

    print(f"Wall is at x={wall_x}, height={wall_height}")

    # Player chooses coefficients
    a = float(input("Enter a: "))
    b = float(input("Enter b: "))
    c = float(input("Enter c: "))

    x_vals = np.linspace(0, 10, 100)
    y_vals = a * x_vals**2 + b * x_vals + c

    # Check if clears the wall
    clears = np.interp(wall_x, x_vals, y_vals) > wall_height

    plt.plot(x_vals, y_vals, label="Projectile Path")
    plt.axvline(wall_x, color='red', linestyle='--', label="Wall")
    plt.axhline(wall_height, color='orange', linestyle=':', label="Wall Height")
    plt.ylim(0, max(y_vals) + 2)
    plt.legend()
    plt.show()

    if clears:
        print("âœ… You cleared the wall!")
    else:
        print(" ")

projectile_game()
```
