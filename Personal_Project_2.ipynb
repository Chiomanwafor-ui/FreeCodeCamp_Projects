```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def load_csv():
    print("Choose how to load the CSV file:")
    print("1. Upload from local computer")
    print("2. Enter URL")
    print("3. Use predefined URL")
    choice = input("Enter 1, 2, or 3: ")

    if choice == '1':
        # For Jupyter or Colab, upload file interactively
        try:
            from google.colab import files
            uploaded = files.upload()
            # Assuming the uploaded file is named 'Fintech_user.csv'
            filename = "Fintech_user.csv"
            if filename not in uploaded:
                print(f"Error: '{filename}' not found among uploaded files.")
                return None
            print(f"Uploaded file: {filename}")
            df = pd.read_csv(filename)
        except ImportError:
            print("Local upload works only in Jupyter or Colab.")
            # Correcting the file path using a raw string
            filepath = r"C:\Users\CHIOMA NWAFOR\Documents\fintech_data.csv"
            try:
                df = pd.read_csv(filepath)
                print(df.head())
            except FileNotFoundError:
                print(f"Error: File not found at {filepath}")
                return None
            except Exception as e:
                print(f"An error occurred while reading the file: {e}")
                return None


    elif choice == '2':
        url = input("Enter the CSV URL: ")
        try:
            df = pd.read_csv(url)
        except Exception as e:
            print(f"An error occurred while reading the URL: {e}")
            return None


    elif choice == '3':
        url = "https://fintech_user/data/.csv" # This URL seems incorrect, please replace with a valid URL if needed
        print(f"Using predefined URL: {url}")
        try:
            df = pd.read_csv(url)
        except Exception as e:
            print(f"An error occurred while reading the predefined URL: {e}")
            return None


    else:
        print("Invalid choice. Please enter 1, 2, or 3.")
        return None

    return df

def explore_data(df):
    if df is None:
        return []
    print("\nDataframe Headings:")
    print(df.columns.tolist())

    print("\nFirst two rows:")
    print(df.head(2))

    cols = df.columns.tolist()
    return cols

def choose_and_plot(df, cols):
    if df is None or not cols:
        print("No data available to plot.")
        return

    print("\nAvailable columns:")
    for i, col in enumerate(cols):
        print(f"{i}: {col}")

    cols_to_plot_input = input("Enter 1 or 2 column indices separated by comma (e.g. 0 or 0,1): ")
    cols_to_plot = [c.strip() for c in cols_to_plot_input.split(',')]

    if len(cols_to_plot) == 1:
        try:
            col_x_index = int(cols_to_plot[0])
            if 0 <= col_x_index < len(cols):
                col_x = cols[col_x_index]
                x_data = df[col_x].to_numpy()
                y_data = None

                plot_type = input("Plot type? Enter 'line' or 'scatter': ").lower()
                plt.figure(figsize=(8,5))

                if plot_type == 'line':
                    plt.plot(x_data)
                    plt.xlabel(col_x)
                    plt.ylabel('Value')
                    plt.title(f"Line plot of {col_x}")

                elif plot_type == 'scatter':
                    y_data = np.arange(len(x_data))
                    plt.scatter(y_data, x_data)
                    plt.xlabel('Index')
                    plt.ylabel(col_x)
                    plt.title(f"Scatter plot of {col_x} vs index")

                else:
                    print("Invalid plot type, defaulting to line plot.")
                    plt.plot(x_data)
                    plt.xlabel(col_x)
                    plt.ylabel('Value')
                    plt.title(f"Line plot of {col_x}")

                plt.grid(True)
                plt.show()
            else:
                print("Invalid column index.")
        except ValueError:
            print("Invalid input for column index.")

    elif len(cols_to_plot) == 2:
        try:
            col_x_index = int(cols_to_plot[0])
            col_y_index = int(cols_to_plot[1])
            if 0 <= col_x_index < len(cols) and 0 <= col_y_index < len(cols):
                col_x = cols[col_x_index]
                col_y = cols[col_y_index]
                x_data = df[col_x].to_numpy()
                y_data = df[col_y].to_numpy()

                plot_type = input("Plot type? Enter 'line' or 'scatter': ").lower()
                plt.figure(figsize=(8,5))

                if plot_type == 'line':
                    plt.plot(x_data, y_data)
                    plt.xlabel(col_x)
                    plt.ylabel(col_y)
                    plt.title(f"Line plot of {col_y} vs {col_x}")

                elif plot_type == 'scatter':
                    plt.scatter(x_data, y_data)
                    plt.xlabel(col_x)
                    plt.ylabel(col_y)
                    plt.title(f"Scatter plot of {col_y} vs {col_x}")

                else:
                    print("Invalid plot type, defaulting to scatter plot.")
                    plt.scatter(x_data, y_data)
                    plt.xlabel(col_x)
                    plt.ylabel(col_y)
                    plt.title(f"Scatter plot of {col_y} vs {col_x}")

                plt.grid(True)
                plt.show()
            else:
                print("Invalid column index.")
        except ValueError:
             print("Invalid input for column indices.")

    else:
        print("Please enter 1 or 2 columns only.")

def data_graph_explorer():
    df = load_csv()
    if df is not None:
        cols = explore_data(df)
        if cols:
            while True:
                choose_and_plot(df, cols)
                again = input("Do you want to plot again? (y/n): ").lower()
                if again != 'y':
                    print("Exiting Data Graph Explorer.")
                    break
        else:
            print("No columns found in the dataframe.")
    else:
        print("Failed to load data.")


data_graph_explorer()
```
